```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
df = pd.read_csv('BTC_USD_2013-10-01_2021-05-18-CoinDesk.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Closing Price (USD)'].plot(figsize=(20, 8));
```

```python
df['SqrtClosing Price (USD)'] = np.sqrt(df['Closing Price (USD)'])
```

```python
df['SqrtClosing Price (USD)'].plot(figsize=(20, 8));
```

```python
df['LogClosing Price (USD)'] = np.log(df['Closing Price (USD)'])
```

```python
df['LogClosing Price (USD)'].plot(figsize=(20, 8));
```

```python
data, lam = boxcox(df['Closing Price (USD)'])
```

```python
lam
```

```python
df['BoxCoxClosing Price (USD)'] = data
df['BoxCoxClosing Price (USD)'].plot(figsize=(20, 8));
```
